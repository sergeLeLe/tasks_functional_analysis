15) Дано числовое уравнение, т.е. уравнение в одной вещественной неизвестной.
1. Преобразовать уравнение к виду, пригодному для применения принципа сжимающих операторов.
2. Методом простых итераций найти приближенное решение с точностью $10^{-5}$,используя априорную и апостериорную оценки числа итераций

$$ \frac{\pi x}{2} + \ln(1+x^2) = 2 \pi + arcctg(\pi x)$$

```python
import numpy as np
import math as mat
import matplotlib.pyplot as plt
%matplotlib inline

x0 = 0
print('x0 = ',x0)

x1 = x0 - (2/(3*mat.pi + 2))*((mat.pi*x0)/2 + np.log(1+x0**2) - 2*mat.pi - (mat.pi/2 - np.arctan(mat.pi*x0)))
print('x1 = ', x1)

#коэф. сжатия
a = 1 - (mat.pi - 2)/(3*mat.pi + 2)
print('a =',a)#0.867

#точность
e = 10**(-4)
print ('e = ', e)

#расстояние
p = np.abs(x0 - x1)
print('p = ', p)

#априорная оценка числа итераций
napr = np.around(np.log((e*(1-a))/p)/(np.log(a))) + 1 #log_a(value)
print('napr = ', napr)

xn_1 = x0
xn = 0

#коэф. для условия
comparable = a/(1-a)

#апостериорная оценка
count = 0
while True:
    count += 1
    xn = xn_1 - (2/(3*mat.pi + 2))*((mat.pi*xn_1)/2 + np.log(1+xn_1**2) - 2*mat.pi - (mat.pi/2 - np.arctan(mat.pi*xn_1)))
    if (comparable*np.abs(xn - xn_1)) <= e:
        print('x_',count,' = ',xn)
        break
    xn_1 = xn

x = np.arange(-5, 10, 0.1)
y1 = x
y2 = [xn_1 - (2/(3*mat.pi + 2))*((mat.pi*xn_1)/2 + np.log(1+xn_1**2) - 2*mat.pi - (mat.pi/2 - np.arctan(mat.pi*xn_1)))
    for xn_1 in x]
plt.figure(figsize=(15,8))
plt.xlabel('xn_1', fontsize = 20)
plt.ylabel('xn', fontsize = 20)
plt.plot(x, y1,'r--', x, y2)  # построение графика
plt.legend(['y = x', '$f(x) = x$'], fontsize = 20) #легенда
plt.grid()              # включение отображение сетки
plt.show()

```
