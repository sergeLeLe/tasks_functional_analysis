Дана функция $x=x(t)$ в пространстве $L^2(-1;1)$ со стандартным скалярным произведением. Для функции $x$ в этом пространстве треюуется найти:

* многочлен наилучшего приближения первой степени;
* многочлен наилучшего приближения второй степени.

Построить графики функции $x$ и полученных многочленов.
<br>


$$
10)\ x(t) = |t|=
\begin{cases}
   t, & t \geq 0\\
   -t, & t < 0
 \end{cases}
$$


**Найдем две частичные суммы ряда Фурье по системе многочленов Лежандра: $S_1(t)$ и $S_2(t)$.**

Возьмем три первых ортонормированных многочлена Лежандра:

$$
P_0(t) = \frac{1}{\sqrt{2}}, \ P_1(t) = \sqrt{\frac{3}{2}}t, \ P_2(t) = \sqrt{\frac{5}{2}}(\frac{3}{2}t^2 - \frac{1}{2})
$$

Вычислим коэффициенты Фурье для функции $x(t)$:

$$
\begin{array}{lcl} 
    c_0 = (x, P_0) = \int\limits_{-1}^1 x(t) P_0 (t) dt = \frac{1}{\sqrt{2}} (\int\limits_{-1}^{0} -t dt + \int\limits_{0}^{1} t dt) = \frac{1}{\sqrt{2}} \\
    \\
    c_1 = (x, P_1) = \int\limits_{-1}^1 x(t) P_1 (t) dt = \sqrt{\frac{3}{2}} (\int\limits_{-1}^{0} -t^2 dt + \int\limits_{0}^{1} t^2 dt) = 0\\
    \\
    c_2 = (x, P_2) = \int\limits_{-1}^1 x(t) P_2 (t) dt = \sqrt{\frac{5}{2}} (\int\limits_{-1}^{0} -t (\frac{3}{2}t^2 - \frac{1}{2}) dt + \int\limits_{0}^{1} t (\frac{3}{2}t^2 - \frac{1}{2}) dt) = \frac{\sqrt{\frac{5}{2}}}{4}
\end{array}
$$



Найдем для функции $x(t)$ многочлены наилучшего приближения первой и второй степени:

$$
\begin{array}{lcl} 
    S_1(t) = c_0 P_0(t) + c_1 P_1(t) = \frac{1}{\sqrt{2}} \cdot \frac{1}{\sqrt{2}} + 0 = \frac{1}{2} \\
    \\
    S_2(t) = c_0 P_0(t) + c_1 P_1(t) + c_2 P_2(t) = \frac{1}{2} + \frac{\sqrt{\frac{5}{2}}}{4} \cdot \sqrt{\frac{5}{2}}(\frac{3}{2}t^2 - \frac{1}{2}) = \frac{15}{16}t^2 + \frac{3}{16}\\
\end{array}
$$

**Построим графики частичных сумм $S_1$ и $S_2$**

```python
import numpy as np
from sympy import *
import matplotlib.pyplot as plt
%matplotlib inline
init_printing()

x = np.arange(-1, 1.1, 0.1)

f,t = symbols('f t')
f = sqrt(t**2)

s_1, s_2 = symbols('s_1 s_2')
s_1 = 1/2*t**0
s_2 = (15/16)*t**2 + 3/16

y = [f.subs(t,val) for val in x]
y1 = [s_1.subs(t, val) for val in x]
y2 = [s_2.subs(t, val) for val in x]

plt.figure(figsize=(15,8))
plt.xlabel('t', fontsize = 20)
plt.ylabel('f(t)', fontsize = 20)
plt.plot(x, y, x, y1, 'r--')
plt.legend(['x(t)', 'S_1(t)'], fontsize = 20)
plt.grid()
plt.show()

plt.figure(figsize=(15,8))
plt.xlabel('t', fontsize = 20)
plt.ylabel('f(t)', fontsize = 20)
plt.plot(x, y, x, y2, 'r--')
plt.legend(['x(t)', 'S_2(t)'], fontsize = 20)
plt.grid()
plt.show()
```
