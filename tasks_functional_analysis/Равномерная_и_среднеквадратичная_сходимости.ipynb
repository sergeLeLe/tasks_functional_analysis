## а) $x_n(t) = \frac{n^2 t}{e^{nt}}$, $x(t) = 0$, $[a,b]=[0,4]$

```python
from sympy import *
import sympy as sym
init_printing()

n,t = symbols('n t')

a = Integral((n**4 * t**2)/exp(2*n*t), (t, 0, 4))
Eq(a,a.doit())
```

```python
b = integrate((n**4 * t**2)/exp(2*n*t), (t, 0, 4))
p = Limit(b, n, sym.oo)
Eq(p,p.doit())
```

```python
import matplotlib.pyplot as plt
from matplotlib import mlab
%matplotlib inline 

import numpy as np

t = np.arange(0, 4.1, 0.01)
plt.figure(figsize=(18,10))
for n in range(1,11):
    f = ((float)(n**2)*t)/(np.exp(n*t))
    plt.plot(t, f)
plt.xlabel('t', fontsize = 20)
plt.ylabel('$x_n(t)$', fontsize = 20)
plt.legend(["$x_1(t)$","$x_2(t)$","$x_3(t)$","$x_4(t)$","$x_5(t)$","$x_6(t)$","$x_7(t)$","$x_8(t)$","$x_9(t)$","$x_{10}(t)$"], fontsize = 20)
plt.grid()
plt.show()
```

## б) $x_n(t) = {3t}^n{1-t}^n $, $x(t) = 0$, $[a,b]=[0,1]$

```python
t = np.arange(-2, 3, 0.001)
plt.figure(figsize=(18,10))
for n in range(1,11):
    f = [((3*x) - 3*(x**2))**n for x in t]
    plt.plot(t, f)
plt.xlabel('t', fontsize = 20)
plt.ylabel('$x_n(t)$', fontsize = 20)
plt.legend(["$x_1(t)$","$x_2(t)$","$x_3(t)$","$x_4(t)$","$x_5(t)$","$x_6(t)$","$x_7(t)$","$x_8(t)$","$x_9(t)$","$x_{10}(t)$"], fontsize = 15)
plt.xlim(-0.5,1.5)
plt.ylim(-0.5, 1)
plt.grid()
plt.show()
```
