18) Дано нелинейное уравнение в пространстве непрерывных функций $C[a;b]$.
1. Используя принцип сжимающих операторов, доказать, что данное уравнение имеет единственное решение $x(t) \in C[a;b]$.
2. Методом простых итераций найти приближенное решение этого уравнения с точностью 0.01, используя априорную оценку числа итераций. В качестве ответа предъявить график приближенного решения.

$$
x(t) - \frac{2}{x^2(t)+2} = t, [a;b] = [-2;2]
$$

```python
from sympy import *
import sympy as sym
import math as mat
init_printing()

t,u = symbols('t u')
ur = (t**2)*cos(u/(2*pi)) - sin(7*t)
du = diff(ur, u)
du
```

```python
import numpy as np
import math as mat
import matplotlib.pyplot as plt
from sympy.calculus.util import minimum, maximum
from scipy import optimize
from sympy import solveset, symbols, Interval, Min, Max
%matplotlib inline

x0 = t
print('x0 = ',x0)

x1 = t + 2/((t**2) + 2)
print('x1 = ', x1)

#коэф. сжатия
alfa = (4*np.sqrt(2/3))/(2/3 + 2)**2
print('alfa =', alfa)

#точность
#e = 0.01
e = 10**(-4)
print ('e = ', e)

#расстояние
p = 1
print('p = ', p)

#априорная оценка числа итераций
napr = np.fix(np.log((e*(1-alfa))/p)/(np.log(alfa))) + 1  #log_a(value)
print('napr = ', napr)

xn_1, xn = symbols('xn_1 xn')
xn_1 = t
xn


#коэф. для условия
comparable = alfa/(1-alfa)

#апостериорная оценка
f = symbols('f')
count = 0
x = np.arange(-2, 2, 0.01)
while True:
    count += 1
    xn = t + 2/((xn_1**2) + 2)
    f = abs(xn - xn_1)
    y = [f.subs(t, val) for val in x]
    if (comparable*max(y) <= e):
        print('x_',count,' = ',xn)
        break
    xn_1 = xn
    

x2 = np.arange(-2, 2, 0.01)
y2 = [xn.subs(t, val) for val in x]

xn_1 = t
xn
#априорная
for i in range(1, int(napr + 1)):
    xn = t + 2/((xn_1**2) + 2)
    xn_1 = xn

print('x_apost = ',xn)

x1 = np.arange(-2, 2, 0.01)
y1 = [xn.subs(t, val) for val in x]

plt.figure(figsize=(15,8))
plt.xlabel('t', fontsize = 20)
plt.ylabel('x(t)', fontsize = 20)
plt.plot(x1, y1, x2, y2)
plt.legend(['априор', 'апостериор'], fontsize = 20)
plt.grid()
plt.show()
```
